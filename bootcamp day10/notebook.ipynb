```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
if 'RESULTS' not in globals():
    RESULTS= {}

iris = load_iris()
X = iris.data
y = iris.target
k_values = [1, 3, 5, 7, 9, 11, 15, 20, 25, 30]
k_scores = {}

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X, y)
    y_pred = model.predict(X)
    k_scores[k] = float(f1_score(y, y_pred, average="weighted"))
best_k = int(max(k_scores, key=k_scores.get))
```

## task09

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data
y = iris.target
X_petals = X[:, 2:]

model_tree_2d = DecisionTreeClassifier(max_depth=3, random_state=42)
model_tree_2d.fit(X_petals, y)
x_min, x_max = X_petals[:, 0].min() - 0.5, X_petals[:, 0].max() + 0.5
y_min, y_max = X_petals[:, 1].min() - 0.5, X_petals[:, 1].max() + 0.5

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.01),
    np.arange(y_min, y_max, 0.01)
)

Z = model_tree_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
colors_region  = ["#a8d8ea", "#a8e6cf", "#ffcef3"]  
colors_scatter = ["#1f77b4", "#2ca02c", "#e91e8c"] 

fig_boundaries_tree, ax = plt.subplots(figsize=(10, 7))

ax.contourf(xx, yy, Z, alpha=0.4, colors=colors_region, levels=[-0.5, 0.5, 1.5, 2.5])

for cls, color, name in zip([0, 1, 2], colors_scatter, iris.target_names):
    mask = y == cls
    ax.scatter(
        X_petals[mask, 0], X_petals[mask, 1],
        c=color, label=name, edgecolors="k", linewidths=0.5, s=60
    )

ax.set_xlabel("Petal length (cm)", fontsize=13)
ax.set_ylabel("Petal width (cm)",  fontsize=13)
ax.set_title("Decision Boundaries — Rule-Based Detector (Decision Tree)", fontsize=14, fontweight="bold")
ax.legend(title="Species", fontsize=11)
plt.tight_layout()
plt.show()
```
